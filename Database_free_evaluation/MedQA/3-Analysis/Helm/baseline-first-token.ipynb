```python
import pandas as pd

df = pd.read_csv('test_MEDQA_first_token_predictions.csv')
df['nct_number'] = df['file_path'].str.extract(r'(NCT\d+)')
letters = ['A', 'B', 'C', 'D']
df['Choice'] = [letters[i % len(letters)] for i in range(len(df))]
columns = ['model_1_gold_lp_moment1', 'model_2_gold_lp_moment1','model_3_gold_lp_moment1', 'model_4_gold_lp_moment1','model_6_gold_lp_moment1', 'predicted_label', 'label', 'Choice', 'source_file', 'file_path', 'nct_number']
df=df[columns]
df
```

```python
def get_fact_letter(group):
    # Find position of 'fact' in the group
    fact_position = group[group['label'] == 'fact'].index[0]
    # Get relative position within the group (0,1,2,3)
    relative_position = fact_position - group.index[0]
    # Map position to letter
    letter_map = {0: 'A', 1: 'B', 2: 'C', 3: 'D'}
    return [letter_map[relative_position]] * len(group)

# Create label-Choice column
df['label-Choice'] = df.groupby('nct_number').apply(get_fact_letter).explode().values
df.head(30)
```

## Group each model by NCT taking max

```python
# Extract NCT number and group by it
# df_result = df_1token
df_1 = df
selected_df_1 = df_1.loc[df_1.groupby('nct_number')['model_1_gold_lp_moment1'].idxmax()]
selected_df_1 = selected_df_1.reset_index(drop=True)
model_1 = pd.DataFrame({'Model_1_pred': selected_df_1['Choice'], 'true': selected_df_1['label-Choice']})
model_1.head()
```

```python
# Extract NCT number and group by it
# df_result = df_1token
df_2 = df
selected_df_2 = df_2.loc[df_2.groupby('nct_number')['model_2_gold_lp_moment1'].idxmax()]
selected_df_2 = selected_df_2.reset_index(drop=True)
model_2 = pd.DataFrame({'Model_2_pred': selected_df_2['Choice'], 'true': selected_df_2['label-Choice']})
model_2.head()
```

```python
# Extract NCT number and group by it
# df_result = df_1token
df_3 = df
selected_df_3 = df_3.loc[df_3.groupby('nct_number')['model_3_gold_lp_moment1'].idxmax()]
selected_df_3 = selected_df_3.reset_index(drop=True)
model_3 = pd.DataFrame({'Model_3_pred': selected_df_3['Choice'], 'true': selected_df_3['label-Choice']})
model_3.head()
```

```python
# Extract NCT number and group by it
# df_result = df_1token
df_4 = df
selected_df_4 = df_4.loc[df_4.groupby('nct_number')['model_4_gold_lp_moment1'].idxmax()]
selected_df_4 = selected_df_4.reset_index(drop=True)
model_4 = pd.DataFrame({'Model_4_pred': selected_df_4['Choice'], 'true': selected_df_4['label-Choice']})
model_4.head()
```

```python
# Extract NCT number and group by it
# df_result = df_1token
df_6 = df
selected_df_6 = df_6.loc[df_6.groupby('nct_number')['model_6_gold_lp_moment1'].idxmax()]
selected_df_6 = selected_df_6.reset_index(drop=True)
model_6 = pd.DataFrame({'Model_6_pred': selected_df_6['Choice'], 'true': selected_df_6['label-Choice']})
model_6.head()
```

```python
# Assuming your DataFrames are named model_1, model_2, model_3, model_4, model_6

# Create final DataFrame starting with true column from any model (they're all the same)
final_df = pd.DataFrame({'true': model_1['true']})

# Add prediction columns from each model
final_df['Model_1_pred'] = model_1['Model_1_pred']
final_df['Model_2_pred'] = model_2['Model_2_pred']
final_df['Model_3_pred'] = model_3['Model_3_pred']
final_df['Model_4_pred'] = model_4['Model_4_pred']
final_df['Model_6_pred'] = model_6['Model_6_pred']

# Print sample and info
print("\nFirst few rows of final DataFrame:")
print(final_df.head())

print("\nShape of final DataFrame:", final_df.shape)

# Print value counts for each column
for column in final_df.columns:
    print(f"\nDistribution of {column}:")
    print(final_df[column].value_counts().sort_index())

# Optional: Save to CSV
final_df.to_csv('combined_model_predictions-first-token.csv', index=False)
final_df.head()
```

```python
# Add mode column
final_df['mode_pred'] = final_df[['Model_1_pred', 'Model_2_pred', 'Model_3_pred', 'Model_4_pred', 'Model_6_pred']].mode(axis=1).iloc[:,0]

# Calculate matches with true labels
matches = (final_df['mode_pred'] == final_df['true']).sum()
total = len(final_df)
accuracy = matches / total

# Print results
print("\nMode prediction results:")
print(f"Total samples: {total}")
print(f"Correct predictions: {matches}")
print(f"Accuracy: {accuracy:.4f}")

# Print distribution of mode predictions
print("\nDistribution of mode predictions:")
print(final_df['mode_pred'].value_counts().sort_index())

# Print sample of results
print("\nSample of final DataFrame with mode:")
print(final_df.head())
```
