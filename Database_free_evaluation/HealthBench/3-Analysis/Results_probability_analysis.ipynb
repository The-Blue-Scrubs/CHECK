```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Configuration variables
N_INTERVALS = 5  # Number of intervals for percentile analysis
INTERVAL_SIZE = 100 // N_INTERVALS  # Size of each interval

# Function to analyze fact class by fact probability percentiles
def analyze_fact_class(df, model_name):
    """Analyze fact class prediction accuracy by fact probability percentiles"""
    # Filter to only examples where model predicted fact (probability_fact >= 0.5)
    df_fact_predictions = df[df['probability_fact'] >= 0.5].copy()
    
    # Sort dataframe by fact probability (highest first)
    df_sorted = df_fact_predictions.sort_values('probability_fact', ascending=False).reset_index(drop=True)
    
    # Define percentile ranges based on N_INTERVALS
    percentiles = range(0, 100, INTERVAL_SIZE)
    
    # Store results
    results = []
    
    for p in percentiles:
        # Calculate row indices for this percentile range
        n_rows = len(df_sorted)
        start_idx = int(n_rows * p/100)
        end_idx = int(n_rows * (p + INTERVAL_SIZE)/100)
        
        # Get data for this percentile range
        df_slice = df_sorted.iloc[start_idx:end_idx]
        
        # Get only fact examples (true label = fact)
        fact_examples = df_slice[df_slice['label'] == 'fact']
        
        # Calculate fact class accuracy
        fact_accuracy = 100 * len(fact_examples) / len(df_slice) if len(df_slice) > 0 else np.nan
        
        # Store results
        accuracies = {
            'percentile': p + INTERVAL_SIZE,  # Use the end of the range for plotting
            'percentile_range': f"{100 - (p + INTERVAL_SIZE)}-{100 - p}%",  # Reverse to show confidence levels
            'n_total': len(df_slice),
            'n_fact': len(fact_examples),
            'fact_prob_range': f"{df_slice['probability_fact'].min():.3f}-{df_slice['probability_fact'].max():.3f}" if len(df_slice) > 0 else "N/A",
            f'{model_name}_fact': fact_accuracy
        }
        results.append(accuracies)
    
    return pd.DataFrame(results)

# Function to analyze hallucination class by hallucination probability percentiles
def analyze_hall_class(df, model_name):
    """Analyze hallucination class prediction accuracy by hallucination probability percentiles"""
    # Compute hallucination probability
    df['probability_hall'] = 1 - df['probability_fact']
    
    # Filter to only examples where model predicted hallucination (probability_fact < 0.5)
    df_hall_predictions = df[df['probability_fact'] < 0.5].copy()
    
    # Sort by hallucination probability (highest first)
    df_sorted = df_hall_predictions.sort_values('probability_hall', ascending=False).reset_index(drop=True)
    
    # Define percentile ranges based on N_INTERVALS
    percentiles = range(0, 100, INTERVAL_SIZE)
    
    # Store results
    results = []
    
    for p in percentiles:
        # Calculate row indices for this percentile range
        n_rows = len(df_sorted)
        start_idx = int(n_rows * p/100)
        end_idx = int(n_rows * (p + INTERVAL_SIZE)/100)
        
        # Get data for this percentile range
        df_slice = df_sorted.iloc[start_idx:end_idx]
        
        # Get only hallucination examples (true label = hallucination)
        hall_examples = df_slice[df_slice['label'] == 'hallucination']
        
        # Calculate hallucination class accuracy
        hall_accuracy = 100 * len(hall_examples) / len(df_slice) if len(df_slice) > 0 else np.nan
        
        # Store results
        accuracies = {
            'percentile': p + INTERVAL_SIZE,  # Use the end of the range for plotting
            'percentile_range': f"{100 - (p + INTERVAL_SIZE)}-{100 - p}%",  # Reverse to show confidence levels
            'n_total': len(df_slice),
            'n_hall': len(hall_examples),
            'hall_prob_range': f"{df_slice['probability_hall'].min():.3f}-{df_slice['probability_hall'].max():.3f}" if len(df_slice) > 0 else "N/A",
            f'{model_name}_hall': hall_accuracy
        }
        results.append(accuracies)
    
    return pd.DataFrame(results)

# Function for combined analysis (for the combined plot)
def analyze_combined_accuracy(df, model_name):
    """Analyze combined accuracy across both classes using the same interval structure"""
    # Sort dataframe by fact probability (highest first)
    df_sorted = df.sort_values('probability_fact', ascending=False).reset_index(drop=True)
    df_sorted['probability_hall'] = 1 - df_sorted['probability_fact']
    
    # Define percentile ranges based on N_INTERVALS
    percentiles = range(0, 100, INTERVAL_SIZE)
    
    # Store results
    results = []
    
    for p in percentiles:
        # Calculate row indices for this percentile range
        n_rows = len(df_sorted)
        start_idx = int(n_rows * p/100)
        end_idx = int(n_rows * (p + INTERVAL_SIZE)/100)
        
        # Get data for this percentile range
        df_slice = df_sorted.iloc[start_idx:end_idx]
        
        # Calculate overall accuracy (correct predictions for both classes)
        correct_predictions = (df_slice['predicted_label'] == df_slice['label']).sum()
        total_predictions = len(df_slice)
        overall_accuracy = 100 * correct_predictions / total_predictions if total_predictions > 0 else np.nan
        
        # Also separate by class for additional info
        fact_examples = df_slice[df_slice['label'] == 'fact']
        hall_examples = df_slice[df_slice['label'] == 'hallucination']
        
        # Store results
        accuracies = {
            'percentile': p + INTERVAL_SIZE,  # Use the end of the range for plotting
            'percentile_range': f"{100 - (p + INTERVAL_SIZE)}-{100 - p}%",  # Reverse to show confidence levels
            'n_total': total_predictions,
            'n_fact': len(fact_examples),
            'n_hall': len(hall_examples),
            'prob_range': f"{df_slice['probability_fact'].min():.3f}-{df_slice['probability_fact'].max():.3f}" if len(df_slice) > 0 else "N/A",
            f'{model_name}_combined': overall_accuracy
        }
        results.append(accuracies)
    
    return pd.DataFrame(results)

# Analyze fact class by fact probability percentiles
fact_results_gpt = analyze_fact_class(df_gpt.copy(), 'GPT-4o')
fact_results_ensemble = analyze_fact_class(df.copy(), 'Ensemble')

# Analyze hallucination class by hallucination probability percentiles
hall_results_gpt = analyze_hall_class(df_gpt.copy(), 'GPT-4o')
hall_results_ensemble = analyze_hall_class(df.copy(), 'Ensemble')

# Merge results for each class
fact_combined = fact_results_gpt.merge(
    fact_results_ensemble[['percentile', 'Ensemble_fact']], 
    on='percentile', how='outer'
)

hall_combined = hall_results_gpt.merge(
    hall_results_ensemble[['percentile', 'Ensemble_hall']], 
    on='percentile', how='outer'
)

# Create subplots for better visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Reverse the percentile values for plotting (so first interval = low confidence, last interval = high confidence)
x_values = (100 + INTERVAL_SIZE) - fact_combined['percentile']  # This reverses the order

# Plot 1: Fact class accuracy by fact probability percentiles
ax1.plot(x_values, fact_combined['GPT-4o_fact'], 'b-o', label='GPT-4o', linewidth=2, markersize=6)
ax1.plot(x_values, fact_combined['Ensemble_fact'], 'r-s', label='Ensemble', linewidth=2, markersize=6)
ax1.set_xlabel('Fact Probability Percentile (%)')
ax1.set_ylabel('Fact Class Accuracy (%)')
ax1.set_title('Fact Class Prediction Accuracy\n(by Fact Confidence Level)')
ax1.grid(True, alpha=0.3)
ax1.legend()
ax1.set_xlim(0, 100)
ax1.set_ylim(0, 110)

# Create dynamic x-axis ticks based on N_INTERVALS
x_ticks = list(range(INTERVAL_SIZE, 101, INTERVAL_SIZE))
ax1.set_xticks(x_ticks)
ax1.set_xticklabels([f"{x}" for x in x_ticks])

# Plot 2: Hallucination class accuracy by hallucination probability percentiles
x_values_hall = (100 + INTERVAL_SIZE) - hall_combined['percentile']  # This reverses the order

ax2.plot(x_values_hall, hall_combined['GPT-4o_hall'], 'b-o', label='GPT-4o', linewidth=2, markersize=6)
ax2.plot(x_values_hall, hall_combined['Ensemble_hall'], 'r-s', label='Ensemble', linewidth=2, markersize=6)
ax2.set_xlabel('Hallucination Probability Percentile (%)')
ax2.set_ylabel('Hallucination Class Accuracy (%)')
ax2.set_title('Hallucination Class Prediction Accuracy\n(by Hallucination Confidence Level)')
ax2.grid(True, alpha=0.3)
ax2.legend()
ax2.set_xlim(0, 100)
ax2.set_ylim(0, 110)
ax2.set_xticks(x_ticks)
ax2.set_xticklabels([f"{x}" for x in x_ticks])

# Add sample size annotations for fact class
for i, row in fact_results_gpt.iterrows():
    if not pd.isna(row['GPT-4o_fact']):
        x_pos = (100 + INTERVAL_SIZE) - row['percentile']  # Reverse the x position
        ax1.annotate(f"n={row['n_fact']}", 
                    (x_pos, row['GPT-4o_fact']),
                    xytext=(0, -15), textcoords='offset points',
                    ha='center', fontsize=7, color='blue', alpha=0.7)

# Add sample size annotations for hallucination class
for i, row in hall_results_gpt.iterrows():
    if not pd.isna(row['GPT-4o_hall']):
        x_pos = (100 + INTERVAL_SIZE) - row['percentile']  # Reverse the x position
        ax2.annotate(f"n={row['n_hall']}", 
                    (x_pos, row['GPT-4o_hall']),
                    xytext=(0, -15), textcoords='offset points',
                    ha='center', fontsize=7, color='blue', alpha=0.7)

plt.tight_layout()
plt.show()

# Print the numerical results
print("\nFact Class Results (by Fact Probability Percentiles):")
print("Only examples where model predicted FACT (probability_fact >= 0.5)")
print("Percentile Range | GPT-4o_fact | Ensemble_fact | n_total | n_fact | Fact_Prob_Range")
print("-" * 95)
for _, row in fact_combined.iterrows():
    fact_range = fact_results_gpt.loc[fact_results_gpt['percentile'] == row['percentile'], 'fact_prob_range'].iloc[0]
    percentile_range = fact_results_gpt.loc[fact_results_gpt['percentile'] == row['percentile'], 'percentile_range'].iloc[0]
    n_total = fact_results_gpt.loc[fact_results_gpt['percentile'] == row['percentile'], 'n_total'].iloc[0]
    n_fact = fact_results_gpt.loc[fact_results_gpt['percentile'] == row['percentile'], 'n_fact'].iloc[0]
    print(f"{percentile_range:>15} | {row['GPT-4o_fact']:>10.1f} | {row['Ensemble_fact']:>12.1f} | {n_total:>7} | {n_fact:>6} | {fact_range}")

print("\nHallucination Class Results (by Hallucination Probability Percentiles):")
print("Only examples where model predicted HALLUCINATION (probability_fact < 0.5)")
print("Percentile Range | GPT-4o_hall | Ensemble_hall | n_total | n_hall | Hall_Prob_Range")
print("-" * 95)
for _, row in hall_combined.iterrows():
    hall_range = hall_results_gpt.loc[hall_results_gpt['percentile'] == row['percentile'], 'hall_prob_range'].iloc[0]
    percentile_range = hall_results_gpt.loc[hall_results_gpt['percentile'] == row['percentile'], 'percentile_range'].iloc[0]
    n_total = hall_results_gpt.loc[hall_results_gpt['percentile'] == row['percentile'], 'n_total'].iloc[0]
    n_hall = hall_results_gpt.loc[hall_results_gpt['percentile'] == row['percentile'], 'n_hall'].iloc[0]
    print(f"{percentile_range:>15} | {row['GPT-4o_hall']:>10.1f} | {row['Ensemble_hall']:>12.1f} | {n_total:>7} | {n_hall:>6} | {hall_range}")

# Create combined results from fact and hallucination class results
def combine_class_results(fact_combined, hall_combined, gpt_model_name='GPT-4o', ensemble_model_name='Ensemble'):
    """Combine fact and hallucination class results to get overall accuracy"""
    combined_results = []
    
    # Get unique percentiles (should be the same for both)
    percentiles = sorted(fact_combined['percentile'].unique())
    
    for p in percentiles:
        # Get fact class data for this percentile
        fact_row = fact_combined[fact_combined['percentile'] == p].iloc[0]
        hall_row = hall_combined[hall_combined['percentile'] == p].iloc[0]
        
        # Get corresponding data from original results to get n_total values
        fact_gpt_row = fact_results_gpt[fact_results_gpt['percentile'] == p].iloc[0]
        fact_ens_row = fact_results_ensemble[fact_results_ensemble['percentile'] == p].iloc[0]
        hall_gpt_row = hall_results_gpt[hall_results_gpt['percentile'] == p].iloc[0]
        hall_ens_row = hall_results_ensemble[hall_results_ensemble['percentile'] == p].iloc[0]
        
        # Calculate combined accuracy for GPT-4o
        fact_correct_gpt = (fact_row[f'{gpt_model_name}_fact'] / 100) * fact_gpt_row['n_total']
        hall_correct_gpt = (hall_row[f'{gpt_model_name}_hall'] / 100) * hall_gpt_row['n_total']
        total_predictions_gpt = fact_gpt_row['n_total'] + hall_gpt_row['n_total']
        combined_accuracy_gpt = 100 * (fact_correct_gpt + hall_correct_gpt) / total_predictions_gpt if total_predictions_gpt > 0 else np.nan
        
        # Calculate combined accuracy for Ensemble
        fact_correct_ens = (fact_row[f'{ensemble_model_name}_fact'] / 100) * fact_ens_row['n_total']
        hall_correct_ens = (hall_row[f'{ensemble_model_name}_hall'] / 100) * hall_ens_row['n_total']
        total_predictions_ens = fact_ens_row['n_total'] + hall_ens_row['n_total']
        combined_accuracy_ens = 100 * (fact_correct_ens + hall_correct_ens) / total_predictions_ens if total_predictions_ens > 0 else np.nan
        
        combined_results.append({
            'percentile': p,
            'percentile_range': f"{100 - (p + INTERVAL_SIZE)}-{100 - p}%",  # Reverse to show confidence levels
            'n_total': total_predictions_gpt,  # Should be same for both models
            'n_fact': fact_gpt_row['n_total'],
            'n_hall': hall_gpt_row['n_total'],
            f'{gpt_model_name}_combined': combined_accuracy_gpt,
            f'{ensemble_model_name}_combined': combined_accuracy_ens
        })
    
    return pd.DataFrame(combined_results)

# Create combined results from existing fact and hallucination analyses
results_combined = combine_class_results(fact_combined, hall_combined)

# Create a combined plot showing overall accuracy
plt.figure(figsize=(8, 6))

# Reverse the percentile values for plotting (so first interval = low confidence, last interval = high confidence)
x_values_combined = (100 + INTERVAL_SIZE) - results_combined['percentile']

# Plot lines for combined accuracy
plt.plot(x_values_combined, results_combined['GPT-4o_combined'], 'b-o', label='GPT-4o', linewidth=2, markersize=6)
plt.plot(x_values_combined, results_combined['Ensemble_combined'], 'r-s', label='Ensemble', linewidth=2, markersize=6)

# Customize the plot
plt.xlabel('Confidence Percentile (%)')
plt.ylabel('Overall Accuracy (%)')
# plt.title('Overall Prediction Accuracy vs Confidence Level\n(Combined Performance from Both Class-Specific Analyses)')
plt.grid(True, alpha=0.3)
plt.legend()

# Set axis limits
plt.xlim(20, 100)
plt.ylim(60, 110)

# Add x-axis ticks for each interval
plt.xticks(x_ticks, [f"{x}" for x in x_ticks])

plt.tight_layout()
plt.show()

# Add combined results to output
print("\nCombined Accuracy Results (from Class-Specific Analyses):")
print("Weighted combination of fact and hallucination class performances")
print("Percentile Range | GPT-4o_combined | Ensemble_combined | n_total | n_fact | n_hall")
print("-" * 85)
for _, row in results_combined.iterrows():
    print(f"{row['percentile_range']:>15} | {row['GPT-4o_combined']:>14.1f} | {row['Ensemble_combined']:>16.1f} | {row['n_total']:>7} | {row['n_fact']:>6} | {row['n_hall']:>6}")

print(f"\nConfiguration: {N_INTERVALS} intervals of {INTERVAL_SIZE}% each") 
```
